## ‚ôªÔ∏è 2. High Availability (HA)

* **Self-healing**: Failed pods are automatically restarted or replaced, minimizing downtime ([BuildPiper][3]).
* **Redundancy and load balancing**: Services distribute workload across healthy pods and nodes, keeping apps available during disruptions ([Prioxis][2]).
* **Multi-zone/region deployment**: Spread across availability zones to mitigate individual node or AZ failures ([GeeksAroundGlobe][4]).
