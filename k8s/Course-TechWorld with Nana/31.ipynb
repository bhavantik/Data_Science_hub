## üîß Step-by-Step Guide

### 1. üöÄ Add Prometheus Community Helm Repo

```bash
helm repo add prometheus-community https://prometheus-community.github.io/helm-charts
helm repo update
```

---

### 2. üîç Search Available Charts

```bash
helm search repo prometheus-community
```

Look for `kube-prometheus-stack` ‚Äì it includes Prometheus, Alertmanager, Grafana, and exporters.

---

### 3. üõ† Install Prometheus Operator Stack

Create a new namespace:

```bash
kubectl create namespace monitoring
```

Install the chart:

```bash
helm install prometheus prometheus-community/kube-prometheus-stack \
  --namespace monitoring
```

> This installs:
>
> * Prometheus Operator
> * Prometheus
> * Alertmanager
> * Grafana
> * Node Exporter
> * Kube State Metrics

---

### 4. üì¶ Verify Components

```bash
kubectl get all -n monitoring
```

You should see multiple pods like:

* `prometheus-operator-*`
* `prometheus-prometheus-*`
* `grafana-*`
* `alertmanager-*`
* `kube-state-metrics`
* `node-exporter`

---

### 5. üåê Access Grafana Dashboard

#### Option 1: Port-forward Grafana (local access)

```bash
kubectl port-forward -n monitoring svc/prometheus-grafana 3000:80
```

Now open:
üëâ [http://localhost:3000](http://localhost:3000)

**Default Credentials:**

* Username: `admin`
* Password: `prom-operator` (or get password with below)

```bash
kubectl get secret -n monitoring prometheus-grafana -o jsonpath="{.data.admin-password}" | base64 --decode
```
